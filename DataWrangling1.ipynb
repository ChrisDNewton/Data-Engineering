#### 4.1: Create a new dataframe for just Tuesdays, NB_MID lane, direction = North, time = between 7:00 and 19:00.

```python
T4 = DF[(DF['Flag Text']=='Tuesday') & (DF['Lane Name']=='NB_MID') & (DF['Direction Name']=='North')\
       & (DF['Hour'].between(7, 18, inclusive = True))]
T4.head()
```

```python
T4.isnull().sum()
```

#### 4.2: Find where missing values are in the dataframe for gaps.

```python
#Find where missing values are in the dataframe for gaps and headways.
null_gaps = T4[T4['Gap (s)'].isnull()]
#null_gaps.head()
null_gaps.Hour.value_counts()
```

```python
T4.groupby('Hour')['Gap (s)'].median()
```

```python
#Rename Gap (s) to gaps
T4['Gaps'] = T4['Gap (s)']
```

```python
#This seems to have worked.
T4['Gap (s)'] = T4.groupby('Hour').Gaps.transform(lambda x: x.fillna(x.median()))
```

```python
T4.head()
```

```python
T4['Gap (s)'].isnull().sum()
```

#### 4.3: Find where missing values are in the dataframe for headways.

```python
#Find where missing values are in the dataframe for headways.
null_headways = T4[T4['Headway (s)'].isnull()]
#null_headways
null_headways.Hour.value_counts()
```

```python
null_headways.groupby('Hour')
null_headways
```

```python
#Get the median Headways value for each hour
T4.groupby('Hour')['Headway (s)'].median()
```

```python
#Rename Headway (s) to Headways so can use it in transform
T4['Headways'] = T4['Headway (s)']
```

```python
#Impute missing headway values with median headway value per hour group
T4['Headway (s)'] = T4.groupby('Hour').Headways.transform(lambda x: x.fillna(x.median()))
```
